# מבוא לעבודה עם Pandas

בכימיה אנחנו עובדים הרבה עם טבלאות: תכונות של חומרים, תוצאות ניסוי, כיולים, ספקטרומים מסוכמים, קבועי שיווי־משקל ועוד. טבלה טובה מאפשרת לשאול שאלות מדעיות בצורה ישירה: איזה חומר רותח בטמפרטורה הגבוהה ביותר? אילו חומרים נוזליים בטמפרטורת החדר? האם למשפחה כימית מסוימת יש טווח רתיחה אופייני?

`pandas` היא ספרייה שמיועדת בדיוק לעבודה כזאת. היא לא מחליפה את NumPy, אלא מוסיפה מעליו שכבה נוחה לעבודה עם נתונים שיש להם **שמות עמודות**, **שמות שורות**, ולעיתים גם טיפוסים שונים בעמודות שונות.


## שני מבני נתונים מרכזיים

ב־Pandas נפגוש בעיקר שני מבנים:

- `DataFrame` - טבלה דו־ממדית. יש לה שורות, עמודות ושמות לעמודות.
- `Series` - סדרה חד־ממדית. אפשר לחשוב עליה כעל עמודה אחת מתוך טבלה, או שורה אחת שחולצה מתוך טבלה.

ההבדל הזה חשוב: טבלה היא דו־ממדית ולכן אפשר לבחור בה גם שורה וגם עמודה. סדרה היא חד־ממדית, ולכן אפשר לבחור בה רק לאורך ציר אחד.


## ייבוא הספרייה

מקובל לייבא את Pandas בשם הקיצור `pd`:


```python
import pandas as pd
```

מעכשיו, פונקציות של Pandas ייקראו דרך `pd`. לדוגמה, `pd.read_csv(...)` טוענת קובץ CSV לתוך `DataFrame`. במחברות הבאות נעבוד עם טבלה קטנה של נקודות התכה ורתיחה של תרכובות נפוצות.
