# שבוע 5 — מבני נתונים

בשבוע השישי נלמד על מבני נתונים חשובים נוספים שמרחיבים את יכולות התכנות שלנו,  
ונראה כיצד ניתן **לשמור, לארגן ולשלוף מידע** בצורה גמישה ונוחה.  
נכיר מבני נתונים כמו tuples ומילונים (dictionaries), ונלמד מתי להשתמש בכל אחד מהם.

נוסיף לכך גם עבודה עם קבצים – איך לקרוא נתונים מקובץ טקסט או לשמור אליהם תוצאות,  
וכן איך לעצב מחרוזות בעזרת כלים מתקדמים ליצירת פלט נקי וברור.

בסוף השבוע תדעו:
- להבין את ההבדלים בין מבני נתונים שונים  
- ליצור ולהשתמש ב-tuples ומילונים
