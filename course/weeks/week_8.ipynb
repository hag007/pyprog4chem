# שבוע 8: ויזואליזציה עם Matplotlib ו־NumPy

בשבוע השמיני נעמיק בעולם הוויזואליזציה המדעית ונלמד כיצד להשתמש בספרייה **Matplotlib** כדי להציג נתונים בצורה ברורה, מדויקת ומשמעותית. נראה כיצד לשלב בין Matplotlib לבין **NumPy** לשרטוט גרפים והפקת תובנות חזותיות מתוך נתונים ניסיוניים או סימולציות. נלמד לבנות גרפים בסיסיים ומתקדמים, לשלב שכבות מידע (labels, legends, grids), וליצור ויזואליזציות שמציגות תופעות כימיות בצורה ברורה ואלגנטית.

בסוף השבוע תדעו:
- ליצור גרפים דו־ממדיים עם Matplotlib ולהתאים את העיצוב שלהם
- לשלב בין NumPy ל־Matplotlib כדי להציג נתונים מחושבים
- להוסיף מקראים, כותרות וטקסטים לגרפים מדעיים
- לבחון תופעה כימית דרך גרפים, תוך שימוש חכם בצירים
