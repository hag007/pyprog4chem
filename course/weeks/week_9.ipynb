# שבוע 9: SciPy לחישובים מדעיים בכימיה

בשבוע התשיעי נכיר את **SciPy**, ספרייה שמרחיבה את NumPy עם כלים לחישובים מדעיים מתקדמים. נשתמש בה כדי להתאים מודלים לנתונים, לפתור משוואות דיפרנציאליות, ולנתח נתונים מרובי־ממדים בעזרת PCA.

הדגש השבוע הוא על תבניות עבודה שימושיות במחקר כמותי: מעבר מנתונים למודל, סימולציה של מערכת כימית לאורך זמן, והצגה ברורה של מבנה בנתונים מורכבים.

בסוף השבוע תדעו:
- להתאים פונקציות לנתונים ניסיוניים ולבחון את איכות ההתאמה
- לעבוד עם התפלגויות ושגיאות מדידה בעזרת SciPy
- לפתור משוואות דיפרנציאליות שמתארות דינמיקה כימית
- להשתמש ב־PCA כדי לצמצם ממדים ולזהות מבנה בנתונים
